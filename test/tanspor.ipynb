```python
import pandas as pd

# Leia o arquivo CSV
nome_arquivo = '1.csv'
dados = pd.read_csv(nome_arquivo)

# Transponha o DataFrame
dados_transpostos = dados.T

# Salve o DataFrame transposto em um novo arquivo CSV
nome_arquivo_transposto = 'dados_transpostos.csv'
dados_transpostos.to_csv(nome_arquivo_transposto, header=False)
```
